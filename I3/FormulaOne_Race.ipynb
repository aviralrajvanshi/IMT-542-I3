```python
import requests
```

```python
result = requests.get("https://api.openf1.org/v1/laps?session_key=9141&driver_number=63").json()
```

```python
duration_per_lap = [] #creating a list

for lap in result: #creating dictionary for data we want to use (number and duraction) and then placing that in our list
  lap_info = {"lap_number": lap["lap_number"], "lap_duration": lap["lap_duration"]}
  duration_per_lap.append(lap_info)

print(duration_per_lap) #printing values from the list we created
```

```python
duration_per_lap = duration_per_lap[1:] #removing first one due to null value
```

```python
import matplotlib.pyplot as plt

# Extract lap numbers and durations
lap_numbers = [lap['lap_number'] for lap in duration_per_lap]
lap_durations = [lap['lap_duration'] for lap in duration_per_lap]

# Plotting
plt.figure(figsize=(12, 6))
plt.plot(lap_numbers, lap_durations, marker='o')
plt.title('Lap Duration over Laps')
plt.xlabel('Lap Number')
plt.ylabel('Lap Duration (seconds)')
plt.grid(True)
plt.show()
```
