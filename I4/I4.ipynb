```python
# this is access method 1: the manual download and upload of Formula One race data history; acquired data set from Kaggle. Provides insights to the total points teams have acquired each year since 1950 to 2020!
from google.colab import files
import pandas as pd

# Prompt user to upload a CSV file
uploaded = files.upload()

# Process and display the uploaded CSV
for filename in uploaded.keys():
    print(f'File "{filename}" uploaded successfully.')
    df = pd.read_csv(filename)
    display(df)  # Use display() in Colab for a nicer table view
```

```python
#Access method 2: reading data from an API for lord of the rings data. This will give us insight as to which editions and which languages are available for all lord of the rigns books at a specific library. Just in case we want an amazing series to start reading!
import requests
import json

# URL for the OpenLibrary API
url = "https://openlibrary.org/search.json?q=the+lord+of+the+rings"

# Make a GET request
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Parse the JSON
    data = response.json()

    # Pretty-print the JSON
    print(json.dumps(data, indent=2))
else:
    print(f"Failed to fetch data. Status code: {response.status_code}")
```

```python
#Access method 2: reading data from an API for margarita cocktail data. Allows us to gain insights as to how to make different types of margaritas, at a cocktail bar level (in case we want to throw a little party at home!)


import requests
import json

# URL for TheCocktailDB API
url = "https://www.thecocktaildb.com/api/json/v1/1/search.php?s=margarita"

# Make a GET request
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Parse the JSON
    data = response.json()

    # Pretty-print the JSON
    print(json.dumps(data, indent=2))
else:
    print(f"Failed to fetch data. Status code: {response.status_code}")

```
